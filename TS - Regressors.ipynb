```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_log_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('coin_Bitcoin.csv')
```

```python
df.head(5)
```

```python
plt.plot(df['Date'], df['Close'])
plt.xlabel('Date')
plt.ylabel('Close')
plt.title('Stock Price over Time')
plt.show()
```

```python
df.head(5)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.head(5)
```

```python
df['dayofweek'] = df['Date'].dt.dayofweek
df['quarter'] = df['Date'].dt.quarter
df['month'] = df['Date'].dt.month
df['year'] = df['Date'].dt.year
df['dayofyear'] = df['Date'].dt.dayofyear
df['sin_day'] = np.sin(df['dayofyear'])
df['cos_day'] = np.cos(df['dayofyear'])
```

```python
df.head(5)
```

```python
df = df.drop(['Date','SNo','Name','Symbol'], axis=1)
```

```python
df.head(5)
```

```python
df.year.value_counts()
```

```python
df_train = df[df['year']<2020]
X_train = df_train.drop(['Close'],axis=1)
y_train = df_train['Close']
```

```python
df_test = df[df['year']>=2020]
X_test = df_test.drop(['Close'],axis=1)
y_test = df_test['Close']
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
X_train.shape, X_test.shape
```

### **Random Forest Regressor**

```python
reg = RandomForestRegressor(n_estimators=500)
reg.fit(X_train,y_train)
yhat = reg.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error

rmse = np.sqrt(mean_squared_error(y_test,yhat))
print("RMSE", rmse)
```

### **Linear Regressor**

```python
reg1 = LinearRegression()
reg1.fit(X_train,y_train)
yhat1 = reg1.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error

rmse1 = np.sqrt(mean_squared_error(y_test,yhat1))
print("RMSE", rmse1)
```
